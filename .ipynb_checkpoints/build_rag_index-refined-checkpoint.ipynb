```python
import os
import re
import pdfplumber
from typing import List, Dict
import vertexai
from vertexai.preview.language_models import TextEmbeddingModel, TextEmbeddingInput
import faiss
import numpy as np
import pickle
from google.cloud import storage
from collections import Counter
from statistics import median

DEBUG = False
```

```python
# -----------------------------
# 1. Define target Item metadata and global heuristics
# -----------------------------

# Define target 10-K Items to extract and embed (used for filtering and labeling)
TARGET_ITEMS = {
    "Item 1.": "BUSINESS",
    "Item 1A.": "RISK_FACTORS",
    "Item 1C.": "CYBERSECURITY",
    "Item 7.": "MDA",
    "Item 7A.": "MARKET_RISK"
}

TARGET_ITEM_RE = re.compile(
    r"^\s*Item\s+(1A|1C|7A|1|7)\b",
    re.IGNORECASE
)

ITEM_RE = re.compile(
    r"^\s*Item\s+\d+[A-Z]?\b",
    re.IGNORECASE
)

# -----------------------------
# 2. Detect TOC
# -----------------------------
ITEM5_RE = re.compile(r"^\s*Item\s+5\b", re.IGNORECASE)

def is_toc_page_simple(lines, page_num, max_front_pages=5, min_item_lines=5):
    if page_num > max_front_pages:
        return False

    has_item5 = any(ITEM5_RE.match(l["text"]) for l in lines)
    if not has_item5:
        return False

    item_lines = sum(1 for l in lines if ITEM_RE.match(l["text"]))
    return item_lines >= min_item_lines

# -----------------------------
# 3. Detect Item headers from bold headline-like lines
# -----------------------------
def detect_item_start(line):

    text = line["text"]

    if not ITEM_RE.match(text):
        return None

    if line["bold_ratio"] < 0.6:
        return None

    m = TARGET_ITEM_RE.match(text)

    if m:
        k = m.group(1).upper()
        label = f"Item {k}."
        return label if label in TARGET_ITEMS else "__OTHER_ITEM__"

    return "__OTHER_ITEM__"
    
# -----------------------------
# 4. Assign Item labels to pages (forward propagation)
# -----------------------------
def label_lines_with_items(lines: List[Dict], start_item=None):
    current_item = start_item
    labeled = []

    for line in lines:
        detected = detect_item_start(line)

        if detected in TARGET_ITEMS:
            current_item = detected
        elif detected == "__OTHER_ITEM__":
            current_item = None

        labeled.append({**line, "item": current_item})

    return labeled, current_item

# -----------------------------
# 5. Low-level text line reconstruction and headline detection
# -----------------------------
END_PUNCT_RE = re.compile(r"[\.!?;:]\s*$")

def norm_line(s: str) -> str:
    s = s.replace("’", "'")
    s = re.sub(r"\s+", " ", s)
    return s

def extract_lines_from_words(page, y_tol=3):

    words = page.extract_words(use_text_flow=True)
    if not words:
        return []

    words = sorted(words, key=lambda w: (w["top"], w["x0"]))

    lines = []
    cur = [words[0]]
    base_top = words[0]["top"]

    for w in words[1:]:
        if abs(w["top"] - base_top) <= y_tol:
            cur.append(w)
        else:
            lines.append(cur)
            cur = [w]
            base_top = w["top"]

    lines.append(cur)

    out = []

    for line_words in lines:

        line_words = sorted(line_words, key=lambda w: w["x0"])

        text = " ".join(w["text"] for w in line_words).strip()

        if not text:
            continue

        # bbox
        x0 = min(w["x0"] for w in line_words)
        x1 = max(w["x1"] for w in line_words)
        top = min(w["top"] for w in line_words)
        bottom = max(w["bottom"] for w in line_words)

        # 提取这一行的 chars
        chars = [
            c for c in page.chars
            if top <= c["top"] <= bottom
        ]

        # bold ratio
        bold_chars = [
            c for c in chars
            if "bold" in c["fontname"].lower()
        ]

        bold_ratio = len(bold_chars) / len(chars) if chars else 0

        out.append({
            "page": page.page_number,
            "text": text,
            "top": top,
            "bottom": bottom,
            "x0": x0,
            "x1": x1,
            "bold_ratio": bold_ratio,
            "break_before": False
        })

    return out[:-1]
    
def mark_paragraph_breaks(lines):
    """
    仅基于页内相邻行之间的垂直空白 gap：
    gap > median_gap  → 认为是新段落起始（break_before=True）
    """
    gaps = [lines[i]["top"] - lines[i-1]["bottom"] for i in range(1, len(lines))]
    base_gap = median(gaps) if gaps else 0.0

    for i in range(1, len(lines)):
        gap = lines[i]["top"] - lines[i-1]["bottom"]
        if gap > (base_gap * 1.1):
            lines[i]["break_before"] = True

    return lines
# -----------------------------
# 6. Filter header
# -----------------------------
PART_RE = re.compile(
    r"^\s*PART\s+[IVXLC]+\b",
    re.IGNORECASE
)

HEADER_ITEM_LIST_RE = re.compile(
    r"^\s*Item\s+\d+[A-Z]?(?:\s*,\s*\d+[A-Z]?)*\s*$",
    re.IGNORECASE
)

PART_RE = re.compile(
    r"^\s*PART\s+[IVXLC]+\b",
    re.IGNORECASE
)

HEADER_ITEM_LIST_RE = re.compile(
    r"^\s*Item\s+\d+[A-Z]?(?:\s*,\s*\d+[A-Z]?)*\s*$",
    re.IGNORECASE
)


def learn_header_templates(pdf, top_k=5, top_band_pct=0.25, min_frac=0.6):
    """
    在主内容区间内学习重复出现的 header 模板。
    自动排除结构型 header：
    PART I / Item 1B,2,3,4
    """

    n = len(pdf.pages)
    end_page = n - 5
    start_page = 5

    if start_page >= end_page:
        return set()

    header_counter = Counter()
    pages_used = 0

    for p_idx in range(start_page, end_page):

        page = pdf.pages[p_idx]
        H = page.height

        lines = extract_lines_from_words(page)

        if not lines:
            continue

        pages_used += 1

        # 只看顶部 band
        top_band = top_band_pct * H

        top_lines = sorted(
            [l for l in lines if l["top"] < top_band],
            key=lambda x: x["top"]
        )[:top_k]

        for l in top_lines:

            text = l["text"].strip()

            # ⭐ 排除结构型 header
            if PART_RE.match(text):
                continue

            if HEADER_ITEM_LIST_RE.match(text):
                continue

            # 学习模板 header
            t = norm_line(text)

            if t:
                header_counter[t] += 1

    if pages_used == 0:
        return set()

    thresh = int(min_frac * pages_used)

    header_templates = {
        t for t, c in header_counter.items()
        if c >= thresh
    }

    return header_templates
    
def drop_headers(lines, page_height, header_templates, top_band_pct=0.25):

    top_band = top_band_pct * page_height

    return [
        l for l in lines
        if not (
            l["top"] <= top_band and (
                PART_RE.match(l["text"].strip())
                or HEADER_ITEM_LIST_RE.match(l["text"].strip())
                or norm_line(l["text"]) in header_templates
            )
        )
    ]
```

```python
if DEBUG:
    pdf_path = "Microsoft 10-K/Microsoft 2024.pdf"

    with pdfplumber.open(pdf_path) as pdf:

        current_item_global = None
    
        for page in pdf.pages:
    
            lines = extract_lines_from_words(page)
            if not lines:
                continue
    
            # 跳过 TOC
            if is_toc_page_simple(lines, page.page_number, max_front_pages=5, min_item_lines=5):
                continue
    
            # 删除 header
            header_templates = learn_header_templates(pdf)
            
            top_band = 0.25 * page.height
            lines = [
                l for l in lines
                if not (
                    l["top"] <= top_band and (
                        PART_RE.match(l["text"].strip())
                        or HEADER_ITEM_LIST_RE.match(l["text"].strip())
                        or norm_line(l["text"]) in header_templates
                    )
                )
            ]
    
            lines = mark_paragraph_breaks(lines)
    
            lines, current_item_global = label_lines_with_items(
                lines,
                start_item=current_item_global
            )
    
            lines = [l for l in lines if l.get("item") in TARGET_ITEMS]
    
            for line in lines:
                marker = "¶" if line["break_before"] else " "
                print(f"{marker} [p{line['page']}] {line['item']} | {line['text']}")
```

```python
def process_single_pdf(pdf_path: str, company: str, year: int):

    embedding_inputs = []

    with pdfplumber.open(pdf_path) as pdf:

        # 1) 学 header 模板（只做一次）
        header_templates = learn_header_templates(pdf)

        current_item_global = None
        paragraph_buffer = []
        current_item = None
        chunk_idx = 0

        def flush(buf, item):
            nonlocal chunk_idx
            if not buf or item not in TARGET_ITEMS:
                return
            text = " ".join(buf).strip()
            if not text:
                return

            safe_item = item.replace(" ", "").replace(".", "")
            chunk_id = f"{company}_{year}_{safe_item}_{chunk_idx}"
            chunk_idx += 1

            embedding_inputs.append({
                "chunk_id": chunk_id,
                "task_type": "RETRIEVAL_DOCUMENT",
                "title": f"{company} {year} {item} #{chunk_idx-1}",
                "content": text,
                "metadata": {
                    "company": company,
                    "year": year,
                    "item": item,
                    "item_type": TARGET_ITEMS[item],
                    "chunk_id": chunk_id
                }
            })

        for page in pdf.pages:

            lines = extract_lines_from_words(page)
            if not lines:
                continue

            # 2) 跳过 TOC
            if is_toc_page_simple(lines, page.page_number, max_front_pages=5, min_item_lines=5):
                continue

            # 3) 清 header
            lines = drop_headers(lines, page.height, header_templates, top_band_pct=0.25)

            # 4) 段落 break
            lines = mark_paragraph_breaks(lines)

            # 5) 跨页 item label
            lines, current_item_global = label_lines_with_items(lines, start_item=current_item_global)

            # 6) 只保留 target items
            lines = [l for l in lines if l.get("item") in TARGET_ITEMS]

            for l in lines:
                item = l["item"]
                text = l["text"].strip()

                # item 变化：flush
                if item != current_item:
                    flush(paragraph_buffer, current_item)
                    paragraph_buffer = []
                    current_item = item

                # 段落开始：flush 上一个段落
                if l.get("break_before") and paragraph_buffer:
                    flush(paragraph_buffer, current_item)
                    paragraph_buffer = []

                paragraph_buffer.append(text)

        # 收尾
        flush(paragraph_buffer, current_item)

    if not embedding_inputs:
        print(f"[WARN] No chunks generated for {company} {year}")
    else:
        print("\n[DEBUG] First chunk content preview:")
        for e in embedding_inputs[:5]:
            print(e["chunk_id"])
            print(e["content"][:200])
            print("-----")
        print("\n-------------------------------\n")

    return embedding_inputs
```

```python
# Process all pdfs
all_embedding_inputs = []

COMPANY = ["Alphabet", "Amazon", "Microsoft", "Oracle"]
YEARS = range(2020, 2026)
PDF_TEMPLATE = "{company} 10-K/{company} {year}.pdf"

for company in COMPANY:
    for year in YEARS:
        pdf_file = PDF_TEMPLATE.format(company=company, year=year)

        if not os.path.exists(pdf_file):
            print(f"Missing file: {pdf_file}, skipping.")
            continue

        print(f"Processing {pdf_file}...")
        year_embeddings = process_single_pdf(
            pdf_path=pdf_file,
            company=company,
            year=year
        )

        all_embedding_inputs.extend(year_embeddings)

print(f"Total chunks prepared: {len(all_embedding_inputs)}")
```

```python
chunks_text = [
    {
        "chunk_id": item["chunk_id"],
        "title": item["title"],
        "content": item["content"]
    }
    for item in all_embedding_inputs
]

with open("chunks_text.pkl", "wb") as f:
    pickle.dump(chunks_text, f)

chunks_metadata = [
    item["metadata"] for item in all_embedding_inputs
]

with open("chunks_metadata.pkl", "wb") as f:
    pickle.dump(chunks_metadata, f)

print("Saved chunks_text.pkl and chunks_metadata.pkl")
```

```python
# Initialize Vertex AI context and load the text embedding model for retrieval
vertexai.init(
    project="qc2360-ieor4526-fall2025",
    location="us-central1"
)
model = TextEmbeddingModel.from_pretrained("text-embedding-005")
```

```python
import time

# Generate vector embeddings for each paragraph-level chunk
embedded_chunks = []

batch_size = 32
max_retries = 5

# 先过滤空内容
valid_items = []
for item in all_embedding_inputs:
    content = item["content"].strip()
    if content:
        valid_items.append(item)

for start in range(0, len(valid_items), batch_size):
    batch = valid_items[start:start + batch_size]

    emb_inputs = [
        TextEmbeddingInput(
            task_type="RETRIEVAL_DOCUMENT",
            title=item["title"],
            text=item["content"].strip()
        )
        for item in batch
    ]

    embeddings = None
    for retry in range(max_retries):
        try:
            embeddings = model.get_embeddings(emb_inputs)
            break
        except Exception as e:
            wait_time = 2 ** retry
            print(f"[WARN] Batch {start // batch_size + 1} failed (retry {retry + 1}/{max_retries}): {e}")
            time.sleep(wait_time)

    if embeddings is None:
        print(f"[SKIP] Batch {start // batch_size + 1} failed after {max_retries} retries.")
        continue

    for item, embedding in zip(batch, embeddings):
        embedded_chunks.append({
            "vector": np.array(embedding.values, dtype="float32"),
            "chunk_id": item["chunk_id"],
            "metadata": item["metadata"],
            "title": item["title"],
            "content": item["content"].strip()
        })

    print(f"Processed batch {start // batch_size + 1} / {(len(valid_items) + batch_size - 1) // batch_size}")

# Report total number of successfully embedded chunks
print(f"Generated embeddings: {len(embedded_chunks)}")
```

```python
if not embedded_chunks:
    raise ValueError("embedded_chunks is empty. No embeddings available to build FAISS index.")

# Stack all embedding vectors into a float32 matrix
X = np.vstack([c["vector"] for c in embedded_chunks]).astype("float32")

# Initialize a FAISS index for similarity search using L2 distance
dim = X.shape[1]
index = faiss.IndexFlatL2(dim)

# Add all embeddings to the FAISS index
index.add(X)

# Verify the number of vectors stored in the index
print("FAISS index size:", index.ntotal)
print("Embedding dimension:", dim)
```

```python
# Persist the FAISS index together with aligned chunk identifiers and metadata
with open("financial_reports_faiss.pkl", "wb") as f:
    pickle.dump({
        "index": index,  # FAISS vector index for similarity search
        "chunk_ids": [c["chunk_id"] for c in embedded_chunks],  # Mapping from index position to chunk ID
        "metadata": [c["metadata"] for c in embedded_chunks],  # Chunk-level metadata (company, year, item, etc.)
        "titles": [c["title"] for c in embedded_chunks]         # Human-readable titles for retrieved chunks
    }, f)
```
