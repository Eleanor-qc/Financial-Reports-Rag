```python
import os
import re
import pdfplumber
from typing import List, Dict
import vertexai
from vertexai.preview.language_models import TextEmbeddingModel, TextEmbeddingInput
import faiss
import numpy as np
import pickle
from google.cloud import storage
```

```python
# -----------------------------
# 1. Define target Item metadata and global heuristics
# -----------------------------

# Define target 10-K Items to extract and embed (used for filtering and labeling)
TARGET_ITEMS = {
    "Item 1.": "BUSINESS",
    "Item 1A.": "RISK_FACTORS",
    "Item 1C.": "CYBERSECURITY",
    "Item 7.": "MDA",
    "Item 7A.": "MARKET_RISK"
}

# Only used to skip early Table of Contents pages
SIGNATURE_PATTERN = re.compile(
    r"\bsignature(s)?\b",
    re.IGNORECASE
)

# -----------------------------
# 2. Detect Item headers from bold headline-like lines
# -----------------------------
def detect_item_start(text: str) -> str | None:
    """
    Detect the start of a 10-K Item section using keyword heuristics
    applied only to bold, headline-like lines.
    """
    t = text.lower()

    # Item 1 – Business
    if "item" in t and "business" in t:
        return "Item 1."

    # Item 1A – Risk Factors
    if "item 1a" in t and "factor" in t:
        return "Item 1A."

    # Item 1C – Cybersecurity
    if "item" in t and "cybersecurity" in t:
        return "Item 1C."

    # Item 7 – MD&A
    if "item" in t and "discussion" in t:
        return "Item 7."

    # Item 7A – Market Risk
    if "item" in t and "market risk" in t:
        return "Item 7A."

    # Any other Item header (boundary only, not stored)
    if "item" in t:
        return "__OTHER_ITEM__"
    # if "item" in t:
    #     print("[DEBUG __OTHER_ITEM__ DETECTED]")
    #     print("TEXT:", repr(text))
    #     return "__OTHER_ITEM__"

    return None

# -----------------------------
# 3. Assign Item labels to pages (forward propagation)
# -----------------------------
def label_lines_with_items(page: List[Dict]) -> List[Dict]:
    """
    Assign Item labels to each extracted text line using forward
    propagation from detected Item headers.
    """
    current_item = None
    labeled = []

    # Only attempt Item detection on bold lines
    for line in page:
        line_text = line["text"]
        detected = None
        if line.get("is_bold"):
            detected = detect_item_start(line_text)

        # Target Item starts
        if detected in TARGET_ITEMS:
            current_item = detected
        # Non-target Item header: end current Item
        elif detected == "__OTHER_ITEM__":
            current_item = None

        labeled.append({
            **line,
            "item": current_item
        })

    return labeled

# -----------------------------
# 4. Group text by Item
# -----------------------------
def collect_item_text(pages: List[Dict]) -> Dict[str, str]:
    """
    Aggregate labeled text lines by Item while preserving table boundaries.
    (Kept for debugging and analysis; not used in embedding construction.)
    """
    item_text = {}
    in_table = False

    for line in pages:
        item = line.get("item")
        if item not in TARGET_ITEMS:
            in_table = False
            continue

        item_text.setdefault(item, "")

        if line.get("is_table_like"):
            if not in_table:
                item_text[item] += "\n[TABLE START]\n"
                in_table = True
            item_text[item] += line["text"] + "\n"
        else:
            if in_table:
                item_text[item] += "[TABLE END]\n"
                in_table = False
            item_text[item] += line["text"] + "\n"

    return item_text

# -----------------------------
# 5. Low-level text line reconstruction and headline detection
# -----------------------------
def extract_lines_from_page(page, y_tolerance=2.0):
    """
    Reconstruct visual text lines by clustering characters with similar
    vertical positions on the PDF page.
    """
    lines = []

    for char in sorted(page.chars, key=lambda c: c["top"]):
        placed = False
        for line in lines:
            if abs(char["top"] - line["top"]) < y_tolerance:
                line["chars"].append(char)
                placed = True
                break
        if not placed:
            lines.append({
                "top": char["top"],
                "chars": [char]
            })

    return lines

def is_headline_line(line, bold_ratio_threshold=0.75):
    """
    Heuristically identify whether a reconstructed line is a section
    headline based on the proportion of bold characters.
    """
    chars = line["chars"]
    if not chars:
        return False

    def is_bold_char(c):
        fname = c["fontname"].lower()
        return any(
            kw in fname
            for kw in [
                "bold",
                "semibold",
                "semi-bold",
                "medium",
                "heavy"
            ]
        )

    bold_count = sum(1 for c in chars if is_bold_char(c))
    return bold_count / len(chars) >= bold_ratio_threshold

def get_line_text(line):
    """
    Concatenate all characters in a reconstructed line into plain text.
    """
    return "".join(c["text"] for c in line["chars"]).strip()

def debug_item_boundaries(pages, target_items=(1, 7), head_n=3):
    """
    Debug utility to print the first few lines and final line of each
    detected Item section to verify boundary correctness.
    """
    from collections import defaultdict

    normalized_targets = []
    for t in target_items:
        t = str(t).strip().upper()
        if not t.startswith("ITEM"):
            t = f"Item {t}."
        normalized_targets.append(t)
    
    buffers = defaultdict(list)

    for p in pages:
        item = p.get("item")
        text = p.get("text")

        for target in target_items:
            if item == target:
                buffers[target].append(p)
            elif buffers[target]:
                # Item just ended
                print(f"\n=== {target} START (first {head_n} lines) ===")
                for line in buffers[target][:head_n]:
                    print(f"[p{line['page']}] {line['text']}")

                last = buffers[target][-1]
                print(f"\n=== {target} END (last line) ===")
                print(f"[p{last['page']}] {last['text']}")
                buffers[target] = []

    # Handle case where document ends inside an item
    for target, lines in buffers.items():
        if lines:
            print(f"\n=== {target} START (first {head_n} lines) ===")
            for line in lines[:head_n]:
                print(f"[p{line['page']}] {line['text']}")

            last = lines[-1]
            print(f"\n=== {target} END (last line) ===")
            print(f"[p{last['page']}] {last['text']}")

# -----------------------------
# 6. Parse PDF, build paragraph-level chunks, and prepare embeddings
# -----------------------------
def process_single_pdf(pdf_path: str, company: str, year: int):
    """
    Parse a single 10-K PDF, detect Item sections, segment text into
    paragraph-level chunks, and prepare embedding payloads with metadata.
    """
    texts = []

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            raw_text = page.extract_text()
            if not raw_text:
                continue

            # Skip TOC in early pages (first 10 pages)
            if i < 10 and SIGNATURE_PATTERN.search(raw_text):
                continue

            # Reconstruct visual text lines from PDF character positions
            lines = extract_lines_from_page(page, y_tolerance=2.0)

            for line in lines:
                text = get_line_text(line)
                if not text:
                    continue

                bold = is_headline_line(line)

                # Heuristic to flag table-like lines dominated by numeric content
                digit_ratio = sum(c.isdigit() for c in text) / max(len(text), 1)
                is_table_like = digit_ratio > 0.3 and len(text.split()) > 3

                texts.append({
                    "page": i + 1,
                    "text": text,
                    "is_bold": bold,
                    "is_table_like": is_table_like
                })

    labeled_pages = label_lines_with_items(texts)

    debug_item_boundaries(
    labeled_pages,
    target_items=("Item 1.", "Item 1A.", "Item 1C.", "Item 7.", "Item 7A.")
)

    embedding_inputs = []

    current_item = None
    paragraph_buffer = []

    def flush_paragraph(buf, item):
        """
        Flush accumulated text lines into a single paragraph chunk
        associated with the current Item.
        """
        if not buf or item not in TARGET_ITEMS:
            return None
        text = " ".join(buf).strip()
        if not text:
            return None
        return text

    paragraph_chunks = []

    for p in labeled_pages:
        line = p["text"].strip()
        item = p.get("item")

        if item != current_item:
            flushed = flush_paragraph(paragraph_buffer, current_item)
            if flushed:
                paragraph_chunks.append((current_item, flushed))
            paragraph_buffer = []
            current_item = item
            continue

        if not line:
            flushed = flush_paragraph(paragraph_buffer, current_item)
            if flushed:
                paragraph_chunks.append((current_item, flushed))
            paragraph_buffer = []
            continue

        if current_item in TARGET_ITEMS:
            paragraph_buffer.append(line)

    flushed = flush_paragraph(paragraph_buffer, current_item)
    if flushed:
        paragraph_chunks.append((current_item, flushed))

    for idx, (item, text) in enumerate(paragraph_chunks):
        item_name = TARGET_ITEMS[item]
        chunk_id = f"{company}_{year}_{item}_{idx}"

        embedding_inputs.append({
            "chunk_id": chunk_id,
            "task_type": "RETRIEVAL_DOCUMENT",
            "title": f"{company} {year} {item} #{idx}",
            "content": text,
            "metadata": {
                "company": company,
                "year": year,
                "item": item,
                "item_type": item_name,
                "chunk_id": chunk_id
            }
        })

    if not embedding_inputs:
        print(f"[WARN] No chunks generated for {company} {year}")

    if embedding_inputs:
        print("\n[DEBUG] First chunk content preview:")
        print(embedding_inputs[0]["content"][:300])
        print("\n-------------------------------\n")

    return embedding_inputs

# -----------------------------
# 7. Process all companies and years
# -----------------------------
all_embedding_inputs = []

COMPANY = ["Alphabet", "Amazon", "Microsoft", "Oracle"]
YEARS = range(2020, 2026)
PDF_TEMPLATE = "{company} 10-K/{company} {year}.pdf"

for company in COMPANY:
    for year in YEARS:
        pdf_file = PDF_TEMPLATE.format(company=company, year=year)

        if not os.path.exists(pdf_file):
            print(f"Missing file: {pdf_file}, skipping.")
            continue

        print(f"Processing {pdf_file}...")
        year_embeddings = process_single_pdf(
            pdf_path=pdf_file,
            company=company,
            year=year
        )

        all_embedding_inputs.extend(year_embeddings)

print(f"Total chunks prepared: {len(all_embedding_inputs)}")

# -----------------------------
# 8. Persist chunk text and metadata for later retrieval
# -----------------------------
chunks_text = [
    {
        "chunk_id": item["chunk_id"],
        "title": item["title"],
        "content": item["content"]
    }
    for item in all_embedding_inputs
]

with open("chunks_text.pkl", "wb") as f:
    pickle.dump(chunks_text, f)

chunks_metadata = [
    item["metadata"] for item in all_embedding_inputs
]

with open("chunks_metadata.pkl", "wb") as f:
    pickle.dump(chunks_metadata, f)

print("Saved chunks_text.pkl and chunks_metadata.pkl")
```

```python
# Initialize Vertex AI context and load the text embedding model for retrieval
vertexai.init(
    project="qc2360-ieor4526-fall2025",
    location="us-central1"
)
model = TextEmbeddingModel.from_pretrained("text-embedding-005")
```

```python
# Generate vector embeddings for each paragraph-level chunk
embedded_chunks = []

for item in all_embedding_inputs:
    # Clean and validate chunk content
    content = item["content"].strip()
    if not content:
        continue

    # Build a retrieval-oriented embedding input (title provides additional context)
    emb_input = TextEmbeddingInput(
        task_type="RETRIEVAL_DOCUMENT",
        title=item["title"],
        text=content
    )

    # Call the embedding model and extract the embedding vector
    embedding = model.get_embeddings([emb_input])[0]

    # Store embedding vector together with chunk identifiers and metadata
    embedded_chunks.append({
        "vector": embedding.values,
        "chunk_id": item["chunk_id"],
        "metadata": item["metadata"],
        "title": item["title"]
    })

# Report total number of successfully embedded chunks
print(f"Generated embeddings: {len(embedded_chunks)}")
```

```python
# Initialize a FAISS index for similarity search using L2 distance
dim = len(embedded_chunks[0]["vector"])
index = faiss.IndexFlatL2(dim)

# Stack all embedding vectors into a float32 matrix as required by FAISS
X = np.array([c["vector"] for c in embedded_chunks]).astype("float32")

# Add all embeddings to the FAISS index for similarity search
index.add(X)

# Verify the number of vectors stored in the index
print("FAISS index size:", index.ntotal)
```

```python
# Persist the FAISS index together with aligned chunk identifiers and metadata
with open("financial_reports_faiss.pkl", "wb") as f:
    pickle.dump({
        "index": index,  # FAISS vector index for similarity search
        "chunk_ids": [c["chunk_id"] for c in embedded_chunks],  # Mapping from index position to chunk ID
        "metadata": [c["metadata"] for c in embedded_chunks],  # Chunk-level metadata (company, year, item, etc.)
        "titles": [c["title"] for c in embedded_chunks]         # Human-readable titles for retrieved chunks
    }, f)
```

```python
client = storage.Client(project="qc2360-ieor4526-fall2025")
bucket = client.bucket("qc2360-fall2025-bucket")

# Upload FAISS index
blob = bucket.blob("rag/financial_reports_faiss.pkl")
blob.upload_from_filename("financial_reports_faiss.pkl")

# Upload chunks_text.pkl
blob = bucket.blob("rag/chunks_text.pkl")
blob.upload_from_filename("chunks_text.pkl")

# Upload chunks_metadata.pkl
blob = bucket.blob("rag/chunks_metadata.pkl")
blob.upload_from_filename("chunks_metadata.pkl")
```
